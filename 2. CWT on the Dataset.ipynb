```python
import tensorflow as tf
import numpy as np
import pywt
from google.colab import drive
drive.mount('/content/drive')
```

```python
file = np.load("/content/drive/MyDrive/Colab Notebooks/CWU Optimization/cwru_archive/CWRU_48k_load_1_CNN_data.npz")
file.files
```

```python
raw_data = file["data"]
raw_data.shape
```

```python
raw_data = raw_data.reshape(-1, 1024)
raw_data.shape
```

```python
wavelet_data = np.repeat(np.nan, repeats = 460 * 10 * 32 * 32).reshape(-1, 32, 32)
for i in range(raw_data.shape[0]):
    segment = raw_data[i, :]
    coefs, _ = pywt.cwt(segment, np.arange(start = 1, stop = 2049, step = 32), "morl")
    wavelet_data[i, :, :] = tf.reshape(tf.image.resize(coefs.reshape((64, 1024, 1)), (32, 32)), (32, 32))
    if (i % 100) == 0 and (i != 0):
        print(f"{i} segments processed.")
```

```python
fault_types = ['Ball_007', 'Ball_014', 'Ball_021', 'IR_007', 'IR_014', 'IR_021', 'OR_007','OR_014', 'OR_021', 'Normal']
labels = np.repeat(fault_types, 460)

# Save wavelet data
np.savez('/content/drive/MyDrive/Colab Notebooks/CWU Optimization/cwru_archive/CWRU_48k_load_1_CNN_wavelet_morlet_data', data = wavelet_data, labels = labels)   
```
